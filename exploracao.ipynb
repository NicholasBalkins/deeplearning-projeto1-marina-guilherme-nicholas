## About Dataset

This data was extracted from the 1994 Census bureau database by Ronny Kohavi and Barry Becker (Data Mining and Visualization, Silicon Graphics). A set of reasonably clean records was extracted using the following conditions: ((AAGE>16) && (AGI>100) && (AFNLWGT>1) && (HRSWK>0)). The prediction task is to determine whether a person makes over $50K a year.

## Data Explanation

O Adult Census Income Dataset, também conhecido como Census Income ou Adult dataset, foi originalmente extraído do Censo dos Estados Unidos de 1994. O objetivo é prever se a renda anual de um indivíduo é superior a US$ 50.000 a partir de variáveis demográficas e profissionais.

## Features 

1. **age**
   Idade do indivíduo em anos. (numérica contínua)

2. **workclass**
   Tipo de emprego ou setor em que a pessoa trabalha (ex.: Private, Self-emp-not-inc, Government, etc.). (categórica nominal)

3. **fnlwgt** (*final weight*)
   Peso amostral atribuído ao registro, indicando quantas pessoas no censo aquele indivíduo representa. (numérica contínua, mas muitas vezes pouco útil para modelos de classificação direta).

4. **education**
   Nível educacional nominal (ex.: Bachelors, HS-grad, Masters, Doctorate). (categórica ordinal)

5. **education.num**
   Representação numérica dos anos de estudo correspondentes ao campo *education* (ex.: HS-grad = 9, Bachelors = 13). (numérica discreta)

6. **marital.status**
   Estado civil do indivíduo (ex.: Married-civ-spouse, Never-married, Divorced, Widowed). (categórica nominal)

7. **occupation**
   Ocupação ou profissão (ex.: Tech-support, Craft-repair, Sales, Exec-managerial). (categórica nominal)

8. **relationship**
   Relação familiar dentro do domicílio (ex.: Husband, Wife, Own-child, Not-in-family). (categórica nominal)

9. **race**
   Categoria racial do indivíduo (ex.: White, Black, Asian-Pac-Islander, Amer-Indian-Eskimo, Other). (categórica nominal)

10. **sex**
    Sexo do indivíduo (Male, Female). (categórica binária)

11. **capital.gain**
    Ganhos de capital reportados (em dólares) — ex.: lucros de investimentos. (numérica contínua, com muitos zeros).

12. **capital.loss**
    Perdas de capital reportadas (em dólares). (numérica contínua, também com muitos zeros).

13. **hours.per.week**
    Número médio de horas trabalhadas por semana. (numérica contínua).

14. **native.country**
    País de origem do indivíduo (ex.: United-States, Mexico, Cuba, etc.). (categórica nominal).

15. **income** (*target*)
    Variável alvo: se a renda anual do indivíduo é **<=50K** ou **>50K** dólares. (binária).


### Variáveis Numéricas

* **age** → idade (contínua)
* **fnlwgt** → peso amostral (contínua, muitas vezes não usada)
* **education.num** → anos de estudo (discreta, ordinal)
* **capital.gain** → ganhos de capital (contínua, muitos zeros)
* **capital.loss** → perdas de capital (contínua, muitos zeros)
* **hours.per.week** → horas trabalhadas por semana (contínua)

---

### Variáveis Categóricas

* **workclass** → tipo de emprego
* **education** → nível educacional (ex.: HS-grad, Masters, Doctorate)
* **marital.status** → estado civil
* **occupation** → profissão
* **relationship** → relação familiar
* **race** → raça/etnia
* **sex** → sexo
* **native.country** → país de origem

---

### Variável Alvo (Target)

* **income** → binária (<=50K, >50K)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# ler o arquivo CSV
df = pd.read_csv(f"dataset/adult.csv")
print(df.head())
```

```python
# limpar os dados
df = df.replace("?", np.nan)
# contar valores nulos em cada coluna
print(df.isnull().sum())
```

```python
# Colunas com missing
cols_missing = ["workclass", "occupation", "native.country"]

# Mostrar a moda de cada uma
for col in cols_missing:
    print(f"Moda de {col}: {df[col].mode()[0]}")
```

```python
# Preencher NaN com a moda
for col in cols_missing:
    df[col] = df[col].fillna(df[col].mode()[0])

print("\nValores nulos depois do preenchimento:")
print(df[cols_missing].isnull().sum())
```

```python
# Mostrar distribuição (top 5 categorias) após preenchimento
for col in cols_missing:
    print(f"\nDistribuição em {col}:")
    print(df[col].value_counts().head())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

sns.set(style="whitegrid")

# --- 1. Distribuição do target (income) ---
plt.figure(figsize=(5,4))
sns.countplot(x="income", data=df, order=df["income"].value_counts().index)
plt.title("Distribuição do income (<=50K vs >50K)")
plt.show()

print(f"% de pessoas com income >50K: {100 * (df['income'] == '>50K').mean():.2f}%")    # é feito uma contagem da proporção de pessoas 
                                                                                        # com income >50K através da média booleana
```

```python
# --- 2. Histogramas das variáveis numéricas ---
num_cols = ["age", "hours.per.week", "education.num", "capital.gain", "capital.loss", "fnlwgt"]

for col in num_cols:
    plt.figure(figsize=(6,4))
    sns.histplot(df[col], bins=40, kde=False)
    plt.title(f"Histograma de {col}")
    plt.show()
```

```python
# Versão log para gains/loss (para ver os outliers)
for col in ["capital.gain", "capital.loss"]:
    plt.figure(figsize=(6,4))
    sns.histplot(np.log1p(df[col]), bins=40, kde=False)
    plt.title(f"Histograma de log1p({col})")
    plt.show()
```

```python
# --- 3. Categóricas ---
cat_cols = ["workclass","education","marital.status","occupation","relationship","race","sex","native.country"]

for col in cat_cols:
    plt.figure(figsize=(8,4))
    sns.countplot(y=col, data=df, order=df[col].value_counts().index[:10])  # top 10 para simplificar
    plt.title(f"Distribuição de {col} (Top 10)")
    plt.show()

# Exemplo de cruzamento com income
plt.figure(figsize=(6,4))
sns.countplot(x="sex", hue="income", data=df)
plt.title("Distribuição de income por sexo")
plt.show()

plt.figure(figsize=(10,5))
sns.countplot(x="education", hue="income", data=df, order=df["education"].value_counts().index)
plt.title("Distribuição de income por educação")
plt.xticks(rotation=45, ha="right")
plt.show()
```

```python
# --- 4. Correlação entre numéricas ---
corr = df[num_cols].corr()
plt.figure(figsize=(7,5))
sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm", center=0)
plt.title("Heatmap de correlação (variáveis numéricas)")
plt.show()
```

```python
# --- 5. Resumo estatístico ---
print(df[num_cols].describe().T)
```

```python
# --- race × income ---
plt.figure(figsize=(7,4))
sns.countplot(x="race", hue="income", data=df, order=df["race"].value_counts().index)
plt.title("Distribuição de income por raça")
plt.xticks(rotation=30, ha="right")
plt.show()
```

```python
# --- occupation × income ---
plt.figure(figsize=(12,5))
sns.countplot(x="occupation", hue="income", data=df, order=df["occupation"].value_counts().index)
plt.title("Distribuição de income por ocupação")
plt.xticks(rotation=45, ha="right")
plt.show()
```

```python
# --- hours.per.week × income (boxplot) ---
plt.figure(figsize=(7,4))
sns.boxplot(x="income", y="hours.per.week", data=df)
plt.title("Distribuição de horas trabalhadas por semana vs income")
plt.show()
```
