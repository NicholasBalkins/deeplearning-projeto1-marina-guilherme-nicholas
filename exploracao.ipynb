## About Dataset

This data was extracted from the 1994 Census bureau database by Ronny Kohavi and Barry Becker (Data Mining and Visualization, Silicon Graphics). A set of reasonably clean records was extracted using the following conditions: ((AAGE>16) && (AGI>100) && (AFNLWGT>1) && (HRSWK>0)). The prediction task is to determine whether a person makes over $50K a year.

## Data Explanation

O Adult Census Income Dataset, tamb√©m conhecido como Census Income ou Adult dataset, foi originalmente extra√≠do do Censo dos Estados Unidos de 1994. O objetivo √© prever se a renda anual de um indiv√≠duo √© superior a US$ 50.000 a partir de vari√°veis demogr√°ficas e profissionais.

## Features 

1. **age**
   Idade do indiv√≠duo em anos. (num√©rica cont√≠nua)

2. **workclass**
   Tipo de emprego ou setor em que a pessoa trabalha (ex.: Private, Self-emp-not-inc, Government, etc.). (categ√≥rica nominal)

3. **fnlwgt** (*final weight*)
   Peso amostral atribu√≠do ao registro, indicando quantas pessoas no censo aquele indiv√≠duo representa. (num√©rica cont√≠nua, mas muitas vezes pouco √∫til para modelos de classifica√ß√£o direta).

4. **education**
   N√≠vel educacional nominal (ex.: Bachelors, HS-grad, Masters, Doctorate). (categ√≥rica ordinal)

5. **education.num**
   Representa√ß√£o num√©rica dos anos de estudo correspondentes ao campo *education* (ex.: HS-grad = 9, Bachelors = 13). (num√©rica discreta)

6. **marital.status**
   Estado civil do indiv√≠duo (ex.: Married-civ-spouse, Never-married, Divorced, Widowed). (categ√≥rica nominal)

7. **occupation**
   Ocupa√ß√£o ou profiss√£o (ex.: Tech-support, Craft-repair, Sales, Exec-managerial). (categ√≥rica nominal)

8. **relationship**
   Rela√ß√£o familiar dentro do domic√≠lio (ex.: Husband, Wife, Own-child, Not-in-family). (categ√≥rica nominal)

9. **race**
   Categoria racial do indiv√≠duo (ex.: White, Black, Asian-Pac-Islander, Amer-Indian-Eskimo, Other). (categ√≥rica nominal)

10. **sex**
    Sexo do indiv√≠duo (Male, Female). (categ√≥rica bin√°ria)

11. **capital.gain**
    Ganhos de capital reportados (em d√≥lares) ‚Äî ex.: lucros de investimentos. (num√©rica cont√≠nua, com muitos zeros).

12. **capital.loss**
    Perdas de capital reportadas (em d√≥lares). (num√©rica cont√≠nua, tamb√©m com muitos zeros).

13. **hours.per.week**
    N√∫mero m√©dio de horas trabalhadas por semana. (num√©rica cont√≠nua).

14. **native.country**
    Pa√≠s de origem do indiv√≠duo (ex.: United-States, Mexico, Cuba, etc.). (categ√≥rica nominal).

15. **income** (*target*)
    Vari√°vel alvo: se a renda anual do indiv√≠duo √© **<=50K** ou **>50K** d√≥lares. (bin√°ria).


### Vari√°veis Num√©ricas

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# ler o arquivo CSV
df = pd.read_csv(f"dataset/adult.csv")
print(df.head())
```

```python
# limpar os dados
df = df.replace("?", np.nan)
# contar valores nulos em cada coluna
print(df.isnull().sum())
```

```python
# Colunas com missing
cols_missing = ["workclass", "occupation", "native.country"]

# Mostrar a moda de cada uma
for col in cols_missing:
    print(f"Moda de {col}: {df[col].mode()[0]}")
```

```python
# Preencher NaN com a moda
for col in cols_missing:
    df[col] = df[col].fillna(df[col].mode()[0])

print("\nValores nulos depois do preenchimento:")
print(df[cols_missing].isnull().sum())
```

```python
# Mostrar distribui√ß√£o (top 5 categorias) ap√≥s preenchimento
for col in cols_missing:
    print(f"\nDistribui√ß√£o em {col}:")
    print(df[col].value_counts().head())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

sns.set(style="whitegrid")

# --- 1. Distribui√ß√£o do target (income) ---
plt.figure(figsize=(5,4))
sns.countplot(x="income", data=df, order=df["income"].value_counts().index)
plt.title("Distribui√ß√£o do income (<=50K vs >50K)")
plt.show()

print(f"% de pessoas com income >50K: {100 * (df['income'] == '>50K').mean():.2f}%")    # √© feito uma contagem da propor√ß√£o de pessoas 
                                                                                        # com income >50K atrav√©s da m√©dia booleana
```

```python
# --- 2. Histogramas das vari√°veis num√©ricas ---
num_cols = ["age", "hours.per.week", "education.num", "capital.gain", "capital.loss", "fnlwgt"]

for col in num_cols:
    plt.figure(figsize=(6,4))
    sns.histplot(df[col], bins=40, kde=False)
    plt.title(f"Histograma de {col}")
    plt.show()
```

```python
# Vers√£o log para gains/loss (para ver os outliers)
for col in ["capital.gain", "capital.loss"]:
    plt.figure(figsize=(6,4))
    sns.histplot(np.log1p(df[col]), bins=40, kde=False)
    plt.title(f"Histograma de log1p({col})")
    plt.show()
```

```python
# --- 3. Categ√≥ricas ---
cat_cols = ["workclass","education","marital.status","occupation","relationship","race","sex","native.country"]

for col in cat_cols:
    plt.figure(figsize=(8,4))
    sns.countplot(y=col, data=df, order=df[col].value_counts().index[:10])  # top 10 para simplificar
    plt.title(f"Distribui√ß√£o de {col} (Top 10)")
    plt.show()

# Exemplo de cruzamento com income
plt.figure(figsize=(6,4))
sns.countplot(x="sex", hue="income", data=df)
plt.title("Distribui√ß√£o de income por sexo")
plt.show()

plt.figure(figsize=(10,5))
sns.countplot(x="education", hue="income", data=df, order=df["education"].value_counts().index)
plt.title("Distribui√ß√£o de income por educa√ß√£o")
plt.xticks(rotation=45, ha="right")
plt.show()
```

```python
# --- 4. Correla√ß√£o entre num√©ricas ---
corr = df[num_cols].corr()
plt.figure(figsize=(7,5))
sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm", center=0)
plt.title("Heatmap de correla√ß√£o (vari√°veis num√©ricas)")
plt.show()
```

```python
# --- 5. Resumo estat√≠stico ---
print(df[num_cols].describe().T)
```

```python
# --- race √ó income ---
plt.figure(figsize=(7,4))
sns.countplot(x="race", hue="income", data=df, order=df["race"].value_counts().index)
plt.title("Distribui√ß√£o de income por ra√ßa")
plt.xticks(rotation=30, ha="right")
plt.show()
```

```python
# --- occupation √ó income ---
plt.figure(figsize=(12,5))
sns.countplot(x="occupation", hue="income", data=df, order=df["occupation"].value_counts().index)
plt.title("Distribui√ß√£o de income por ocupa√ß√£o")
plt.xticks(rotation=45, ha="right")
plt.show()
```

```python
# --- hours.per.week √ó income (boxplot) ---
plt.figure(figsize=(7,4))
sns.boxplot(x="income", y="hours.per.week", data=df)
plt.title("Distribui√ß√£o de horas trabalhadas por semana vs income")
plt.show()
```
